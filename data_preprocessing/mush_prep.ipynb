```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_path = '/nmnt/x04-hdd/boris_temp/SGIMC_IMC/mushrooms/'
data = pd.read_csv(data_path + 'mushrooms', sep = ' ', header=None)
data.drop(labels=[data.columns[-1]], axis=1, inplace=True)
```

```python
data.head()
```

```python
data_arr = data.as_matrix()
features_arr = data_arr[:,1:]
atr_arr = data_arr[:,0].astype(dtype='int32')

for i in range(features_arr.shape[0]):
    for j in range(features_arr.shape[1]):
        features_arr[i,j] = features_arr[i,j][:-2]
features_arr = features_arr.astype(dtype='int32')
features_ids = np.unique(features_arr)

R = np.zeros((len(data_arr), len(data_arr)))
X = np.zeros((len(data_arr), len(features_ids)))
Y = np.eye(len(data_arr))

# creating matrix for semi-supervised clustering
for i in range(len(atr_arr)):
    for j in range(len(atr_arr)):
        if atr_arr[i] == atr_arr[j]: R[i,j] = 1
        else: R[i,j] = -1
            
# creating matrix of features
for i, x in enumerate(features_arr):
    for feature in x:
        for j, f in enumerate(features_ids):
            if feature == f: X[i,j] = 1
                
# creating matrix of features
for i, x in enumerate(features_arr):
    for feature in x:
        for j, f in enumerate(features_ids):
            if feature == f: X[i,j] = 1
```

```python
np.save(data_path + 'R.npy', R)
np.save(data_path + 'X.npy', X)
np.save(data_path + 'Y.npy', Y)
```
